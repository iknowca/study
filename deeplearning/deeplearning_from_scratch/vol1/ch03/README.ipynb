가장 왼쪽 노드들을 입력층, 오른쪽 노드들은 출력층, 그외의 층들은 은닉층이라고 한다.

퍼셉트론은 다음과 같은 식과 그림으로 나타난다.

$$
y = \begin{cases}
1, & (b + w_1 x_1 + w_2 x_2) > 0 \\
0, & (b + w_1 x_1 + w_2 x_2) \leq 0
\end{cases}
$$

위의 식을 더 간결하게 나타내면 다음과 같이 나타낼 수 있다.

$$
y = h(b + w_1 x_1 + w_2 x_2)
$$
$$
h(x) = \begin{cases}
1, & x > 0 \\
0, & x \leq 0
\end{cases}
$$

입력신호의 총 합이 $h(x)$를 거쳐 변환된 값이 y의 출력이 된다.

## Activation funcion

입력신호의 총합을 출력신호로 변환하는 함수를 활성화함수라고 한다.
$$
a = b + w_1x_1 + w_2 x_2
$$
$$
y = h(a)
$$

입력신호와 가중치의 곱을 계산하여 편향과 총합을 구하고 이를 $a$라하고 $a$를 함수 $h(a)$를 계산하여 $y$를 출력한다
