$$y = \begin{cases}
0 & (b+w_1x_1 + w_2x_2) \leq 0 \\
1 & (b+w_1x_1 + w_2x_2) > 0 \\
\end{cases}
$$

$$y = h(b + w_1x_1 + w_2x_2)$$
$$
h(x) = \begin{cases}
0 & (x \leq 0) \\
1 & (x > 0) \\
\end{cases}
$$

h(x) is activation function

if $b + w_1x_1 + w_2x_2 = a$, then $y = h(a)$

### step function

```python
def step_function(x):
    if x>0:
        return 1
    else:
        return 0
```

for numpy

```python
def step_function(x):
    y = x > 0
    return y.astype(int)
```

example

```python
import numpy as np
x = np.array([-1.0, 1.0, 2.0])
print(x)

y = x>0
print(y)
```

graph

```python
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```

### sigmoid function


```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y_step_function = step_function(x)
y_sigmoid = sigmoid(x)
plt.plot(x, y_step_function, label="step_function")
plt.plot(x, y_sigmoid, label="sigmoid")
plt.ylim(-0.1, 1.1)
plt.legend()
```
