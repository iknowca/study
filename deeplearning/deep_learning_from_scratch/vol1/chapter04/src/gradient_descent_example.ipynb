```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def numerical_gradient(f, x):
    h = 1e-4
    grad = np.zeros_like(x)

    for idx in range(x.size):
        tmp_val = x[idx]
        x[idx] = tmp_val + h
        fxh1 = f(x)

        x[idx] = tmp_val -h
        fxh2 = f(x)

        grad[idx] = (fxh1 - fxh2) / (2*h)
        x[idx] = tmp_val
    return grad
```

```python
def gradient_descent(f, init_x, lr=10, step_num=100):
    x = init_x
    x_history = [x.copy()]
    for i in range(step_num):
        grad = numerical_gradient(f, x)
        x -= lr * grad
        x_history.append(x.copy())
    return (x, x_history)
```

```python
def func(x):
    return x[0]**2 + x[1]**2
```

```python
init_x = np.array([-3.0, -4.0])
x, history = gradient_descent(func, init_x, lr=0.9, step_num=15)
```

```python
import matplotlib.patches as patches

for i in range(16):
    plt.scatter(*history[i], marker="$"+str(i)+"$")

plt.gca().add_patch(patches.Ellipse((0, 0), 10, 10, fc="none", ec="grey"))
plt.gca().add_patch(patches.Ellipse((0, 0), 5, 5, fc="none", ec="grey"))
plt.gca().add_patch(patches.Ellipse((0, 0), 2.5, 2.5, fc="none", ec="grey"))
plt.gca().add_patch(patches.Ellipse((0, 0), 1, 1, fc="none", ec="grey"))

plt.title(r"lr$=0.9$ step_number$=15$")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```
