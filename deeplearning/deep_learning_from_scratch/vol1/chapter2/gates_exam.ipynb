```python
# visualize OR gate linear function
import matplotlib.pyplot as plt

x00 = [0, 0]
x01 = [0, 1]
x10 = [1, 0]
x11 = [1, 1]

plt.scatter(*x00, color='red')
plt.scatter(*x01, color='blue')
plt.scatter(*x10, color='blue')
plt.scatter(*x11, color='blue')

plt.plot([-0.5, 1], [1, -0.5], color='black')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-0.3, 1.2)
plt.ylim(-0.3, 1.2)
plt.fill_between([-0.5, 1], [1, -0.5], -0.3, color='gray', alpha=0.5)
plt.title('OR Gate Visualization')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.legend(['False', 'True'], loc='upper left')
plt.show()
```

```python
# visualize XOR gate
plt.scatter(*x00, color='red')    # (0,0) -> output 0
plt.scatter(*x01, color='blue')   # (0,1) -> output 1
plt.scatter(*x10, color='blue')   # (1,0) -> output 1
plt.scatter(*x11, color='red')    # (1,1) -> output 0

# Draw decision boundaries for XOR (two lines needed)
plt.plot([-0.3, 0.5], [0.5, 1.2], color='black')
plt.plot([0.2, 1.2], [-0.3, 0.7], color='black')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-0.3, 1.2)
plt.ylim(-0.3, 1.2)
plt.title('XOR Gate Visualization')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.legend(['False', 'True'], loc='upper left')
plt.show()
```
