```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def sigmoid(x, delta=1e-6):
    return 1 / (1 + np.exp(-x+delta))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid function')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```
