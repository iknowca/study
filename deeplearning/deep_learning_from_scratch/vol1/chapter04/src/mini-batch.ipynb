Implementing Cross Entropy Error for Minibatches

```python
def cross_entropy_error(y, t, delta=1e-7):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
        
    batch_size = y.shape[0]
    return -np.sum(t*np.log(y+delta)) /batch_size
```

```python
def cross_entropy_error_with_label(y, t, delta=1e-7):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
        
    batch_size = y.shape[0]
    return -np.sum(t*np.log(y[np.arange(batch_size), t]+delta)) /batch_size
```
